```python
# 🌟 Exercise 2: Descriptive Statistics

#     Task: Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
import scipy.stats as stats
import numpy as np
#     Sample Dataset:
data = [12, 15, 13, 12, 18, 20, 22, 21]
mean_scipy = stats.tmean(data)
median_scipy = np.median(data)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)
print(f"""
mean: {mean_scipy}
median: {median_scipy}
variance: {variance_scipy}
std. dev.: {std_dev_scipy}
      """)
```

```python
# 🌟 Exercise 3: Understanding Distributions

#     Task: Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.
#     Code Example:

# import numpy as np -- don't need to import bc imported below.
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Define distribution parameters
mean = 50
std_dev = 10

# 2. Create a range of x-values (from -4 to +4 standard deviations)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# 3. Calculate the Probability Density Function (PDF)
y = norm.pdf(x, mean, std_dev)

# 4. Plot the distribution
plt.plot(x, y, label=f'Mean={mean}, Std. Dev.={std_dev}', color='blue')
plt.fill_between(x, y, color='blue', alpha=0.2)
plt.title('Normal Distribution (Bell Curve)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# 🌟 Exercise 6: Understanding ANOVA

# Objective: Test understanding of ANOVA and its application.

#     Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.
#     Dataset:
#         fertilizer_1: [5, 6, 7, 6, 5]
#         fertilizer_2: [7, 8, 7, 9, 8]
#         fertilizer_3: [4, 5, 4, 3, 4]
fertilizer_1 = np.array([5, 6, 7, 6, 5])
fertilizer_2 = np.array([7, 8, 7, 9, 8])
fertilizer_3 = np.array([4, 5, 4, 3, 4])
#     Questions:
#     Perform an ANOVA test on the given data. What are the F-value and P-value?
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)

```

```python
#     Based on the P-value, do the fertilizers have significantly different effects on plant growth?
# Yes, the p-value is very small, much less than 0.05, indicating that we can reject the null hypothesis, and there is a statistically significant difference between at least two of the groups; at least two of the fertilizers have significantly different effects on plant growth.
#     Explain what would happen if the P-value were greater than 0.05.
# If the p-value were greater than 0.05, we could not reject the null hypothesis, so we would be unable to declare that the fertilizers have significantly different effects on plant growth.
```

```python
# 🌟 Exercise 8: Correlation Coefficients (Optional)

#     Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
#     Example in code:
import pandas as pd
from scipy.stats import pearsonr
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

corr, _ = pearsonr(data['age'], data['income'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')

from scipy.stats import spearmanr

corr, _ = spearmanr(data['age'], data['income'])
print(f'Spearman Rank Correlation: {corr:.2f}')
```
