```python
#1 Given a list [1, 2, 3, 4], print out all the values in the list one by one.
list = [1,2,3,4]
for i in list:
    print(i)
```

```python
#2 Given a list [1, 2, 3, 4], print out all the values in the list multiplied by 20.

list = [1,2,3,4]
newlist = []
for i in list:
    newlist.append(i*20)
print(newlist)
```

```python
#3 Given a list ["Elie", "Tim", "Matt"],
# return a new list with only the first letter of each name: ["E", "T", "M"].
list = ['Elie','Tim','Matt']
newlist = []
for i in list:
    newlist.append(i[0])
print(newlist)
```

```python
#4 Given a list [1, 2, 3, 4, 5, 6], return a new list with all the even values: [2, 4, 6].
list = [1,2,3,4,5,6]
newlist = [i for i in list if i%2==0]
print(newlist)
```

```python
#5 Given two lists [1, 2, 3, 4] and [3, 4, 5, 6],
# return a new list that contains only the values present in both lists: [3, 4].

a = [1,2,3,4]
b = [3,4,5,6]
c = [i for i in a if i in b]
print(c)
```

```python
#6 Given a list of words ["Elie", "Tim", "Matt"], return a new list with each word
# reversed and in lowercase: ["eile", "mit", "ttam"].

list = ['Elie','Tim','Matt']
newlist = []
for i in list:
    newlist.append(i[::-1].lower())
print(newlist)
```

```python
#7 Given two strings "first" and "third", return a new list of the letters that are present
# in both strings: ["i", "r", "t"].

a = "first"
b = "third"
c = [i for i in a if i in b]
print(c)
```

```python
#8 For all numbers between 1 and 100,
# return a list of the numbers that are divisible by 12: [12, 24, 36, 48, 60, 72, 84, 96].
divby12s = [i for i in range(1,101) if i%12==0]
divby12s
```

```python
#9 Given the string "amazing", return a list with all the vowels removed: ["m", "z", "n", "g"].
mystring = "amazing"
newlist = []
for i in mystring:
    if i in ['a','e','i','o','u']:
        continue
    else:
        newlist.append(i)
newlist
```
