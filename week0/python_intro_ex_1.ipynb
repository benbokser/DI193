```python
humanYears = int(input("How many years ago did you get the cat and dog?"))
if humanYears == 1:
    catYears = 15
    dogYears = 15
elif humanYears == 2:
    catYears = 15+9
    dogYears = 15+9
elif humanYears > 2:
    catYears = 15+9 + 4*(humanYears - 2)
    dogYears = 15+9 + 5*(humanYears - 2)
print(f"[{humanYears},{catYears},{dogYears}]")
```
