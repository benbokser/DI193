```python
# Exercise 3
# Instructions

# Write a script that calculates the number of upper case letters and lower case letters in a string.
def type_count(string):
    uppers = 0
    lowers = 0
    for letter in string:
        if letter.isupper():
            uppers += 1
        if letter.islower():
            lowers += 1
    return uppers, lowers
print(type_count('dkfjsJJJddFFFJ'))
```

```python
# Exercise 4
# Instructions

# Write a function to find the sum of an array without using the built in function:

# >>>my_sum([1,5,4,2])
# >>>12
def my_sum(list):
    result = 0
    for i in list:
        result += i
    return result

print(my_sum([1,5,4,2]))
```

```python
# Exercise 5
# Instructions

# Write a function to find the max number in a list

# >>>find_max([0,1,3,50])
# >>>50

def find_max(list):
    max = 0
    for i in list:
        if i > max:
            max = i
    return max
print(find_max([3,6,3,2,6,8]))

```

```python

# Exercise 6
# Instructions

# Write a function that returns factorial of a number

# >>>factorial(4)
# >>>24
def factorial(num):
    result = 1
    for i in range(1,num+1):
        result = result*i
    return result
print(factorial(24))
print(factorial(7))
```

```python
# Exercise 7
# Instructions

# Write a function that counts an element in a list (without using the count method):
def list_count(list, element):
    result = 0
    for i in list:
        if i == element:
            result += 1
    return result    
print(list_count(['a','a','t','o'],'a'))
# >>>2
```

```python
# Exercise 8
# Instructions

# Write a function that returns the L2-norm (square root of the sum of squares) of the sum of a list:
def norm(list):
    sum_squares = sum([i**2 for i in list])
    result = sum_squares**0.5
    return result
print(norm([1,2,2]))
# >>>3
```

```python
# Exercise 9
# Instructions

# Write a function to find if an array is monotonic (sorted either ascending of descending)

def is_mono(array):
    if array == sorted(array) or array == sorted(array, reverse=True):
        return True
    else:
        return False



print(is_mono([7,6,5,5,2,0]))
# >>>True

print(is_mono([2,3,3,3]))
# >>>True

print(is_mono([1,2,0,4]))
# >>>False
```

```python
# Exercise 10
# Instructions

# Write a function that prints the longest word in a list.
def print_longest_word(list):
    longest_word = ''
    for word in list:
        if len(word) > len(longest_word):
            longest_word = word
    print(longest_word)
print_longest_word(['one','two','three'])
```

```python
# Exercise 11
# Instructions

# Given a list of integers and strings, put all the integers in one list, and all the strings in another one.
def list_allocator(list):
    int_list = []
    str_list = []
    for i in list:
        if type(i) == int:
            int_list.append(i)
        elif type(i) == str:
            str_list.append(i)
    return int_list, str_list
            
print(list_allocator([3,'ad','35','dd',3,6,'dds']))
```

```python
# Exercise 12
# Instructions

# Write a function to check if a string is a palindrome:
def is_palindrome(string):
    for i in range(len(string)//2):
        if string[i] != string[-(i+1)]:
            return False
    return True

print(is_palindrome('radar'))
# >>>True

print(is_palindrome('John'))
# >>>False
print(is_palindrome('sas'))
```

```python
# Exercise 13
# Instructions

# Write a function that returns the amount of words in a sentence with length > k:
def sum_over_k(sentence,k):
    word_list = sentence.split(' ')
    output = 0
    for word in word_list:
        if len(word) > k:
            output += 1
    return output
sentence = 'Do or do not there is no try'
k=2
print(sum_over_k(sentence,k))
# >>>3
```

```python
# Exercise 15
# Instructions

# Write a function that returns common divisors of 2 numbers:
def common_div(num1,num2):
    #find the smaller of the two. initialize an empty list. then start a for loop over the range 
    # from 2 to the smaller number (inclusive), and add each element to the list if it has 0 % for both numbers.
    smaller = min(num1,num2)
    divisors = []
    for i in range(2,smaller+1):
        if num1%i == num2%i == 0:
            divisors.append(i)
    return divisors
print(common_div(10,20))

# >>>common_div(10,20)
# >>>[2,5,10]
```

```python
# Exercise 16
# Instructions

# Write a function that test if a number is prime:
def is_prime(num):
    for i in range(2,num):
        if num%i == 0:
            return False
    return True
print(is_prime(11))
print(is_prime(12))
print(is_prime(8))
print(is_prime(7))
# >>>True
```

```python
# Exercise 17
# Instructions

# Write a function that prints elements of a list if the index and the value are even:
def weird_print(list1):
    evens = []
    for index,element in enumerate(list1):
        if index%2 == element%2 == 0:
            evens.append(element)
    print(evens) 
weird_print([1,2,2,3,4,5])
# >>>[2,4]
```

```python
# Exercise 18
# Instructions

# Write a function that accepts an undefined number of keyworded arguments and return the count of different types:
def type_count(**kwargs):
    num_int = 0
    num_str = 0
    num_float = 0
    num_bool = 0
    for i in kwargs.values():
        if type(i) == int:
            num_int +=1
        elif type(i) == str:
            num_str +=1
        elif type(i) == float:
            num_float +=1
        elif type(i) == bool:
            num_bool +=1
    
    print(f'int: {num_int}, str: {num_str}, float: {num_float}: bool: {num_bool}')
    return num_int, num_str, num_float, num_bool
type_count(a=1,b='string',c=1.0,d=True,e=False)
# >>>int: 1, str:1 , float:1, bool:2
```

```python
# Exercise 19
# Instructions
# Write a function that mimics the builtin .split() method for strings.
def splitter(str1, sep=' '):
    result = []
    current_word = ''
    for index, i in enumerate(str1):
        if i != sep:
            current_word = current_word + i
            if index == len(str1) -1:
                result.append(current_word)
        elif i == sep:
            if current_word != '':
                result.append(current_word)
                current_word = ''
        
    return result
# By default the function uses whitespace but it should be able to take an argument for any character and split with that argument.
print(splitter('This is  a test'))
print(splitter('This/is/another/test/checking/now','/'))
```
