```python
# Exercise 1 : Advanced Matrix Operations
# Instructions
# Create a 5x5 matrix with random numbers and perform the following operations:
# 1. Calculate the eigenvalues and eigenvectors.
# 2. Perform matrix normalization.
import numpy as np

m1 = np.random.randint(0,101, 25).reshape(5,5)
print(f'matrix:{m1}')
eigenvalues, eigenvectors = np.linalg.eig(m1)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

#min-max normalization:
m1_minmax = (m1 - m1.min()) / (m1.max() - m1.min())
print(f'''Matrix after min-max normalization:
{m1_minmax}''')
```

```python
# Exercise 2 : Statistical Distribution Analysis
# Instructions

# Hint: To generate a dataset following a normal distribution, you can use numpy’s np.random.normal() function. For plotting its histogram, matplotlib’s plt.hist() can be used.

# Generate a dataset following a normal distribution and plot its histogram. Then, calculate and interpret its skewness and kurtosis.

import matplotlib.pyplot as plt
import seaborn as sns

normal_data = np.random.normal(0,1,500)
plt.hist(normal_data)
plt.title('Histogram of randomly generated normal data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import skew, kurtosis

# Calculating skewness and kurtosis
skewness = skew(normal_data)
kurtosis_value = kurtosis(normal_data)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis_value)
```

```python
# Exercise 3 : Time Series Forecasting
# Instructions
import numpy as np
# # Generating monthly sales data as a NumPy array
monthly_sales_data = np.random.randint(100, 500, size=12)  # Random sales data for 12 months

# Using a simple NumPy array representing time series data (e.g., monthly sales data), demonstrate how to forecast future data points using linear regression.
months = np.arange(len(monthly_sales_data))

# Use np.polyfit to calculate the linear regression (degree=1)
# Returns [slope, intercept]
slope, intercept = np.polyfit(months, monthly_sales_data, 1)

# Define the Forecast Function: y = mx + b
def predict_sales(month_index):
    return slope * month_index + intercept

# Forecast the next 6 months (Months 12, 13, 14, 15, 16, 17)
future_months = np.arange(12, 18)
forecast_results = np.round(predict_sales(future_months)).astype(int)

print(f"Linear Equation: Sales = {slope:.2f} * Month + {intercept:.2f}")
print(f"Next 6 Months Forecast: {forecast_results[:6]}")
```

```python
# Exercise 4 : Pandas Data Aggregation
# Instructions


import pandas as pd

# Creating a sample Pandas DataFrame
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B'],
    'Sales': [200, 150, 300, 250, 180],
    'Month': ['January', 'February', 'March', 'April', 'May']
}
df = pd.DataFrame(data)

# Using a Pandas DataFrame, demonstrate grouping and aggregation techniques to summarize the data effectively.
df_grouped = df.groupby('Product')['Sales'].agg(['sum', 'mean', 'median','count'])
df_grouped
```

```python
# Exercise 5 : Complex Data Visualization
# Instructions

# Create a multi-plot layout using Matplotlib and plot different types of graphs (e.g., line, scatter, bar) with NumPy-generated data.


import matplotlib.pyplot as plt
import numpy as np

# 1. Generate NumPy Data
x = np.linspace(0, 10, 100)
y_sine = np.sin(x)                               # Line data
scatter_x = np.random.randn(100)                 # Scatter data
scatter_y = np.random.randn(100)
categories = ['Coal', 'Gas', 'Hydro', 'Solar']   # Bar data
values = np.random.randint(20, 100, size=4)
dist_data = np.random.normal(50, 15, 1000)        # Histogram data

# 2. Create a 2x2 Layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Comprehensive Energy Data Analysis (NumPy Powered)', fontsize=16)

# --- Plot 1: Line Graph (Top Left) ---
axs[0, 0].plot(x, y_sine, color='tab:blue', linewidth=2)
axs[0, 0].set_title('Grid Stability (Sine Wave)')
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# --- Plot 2: Scatter Plot (Top Right) ---
colors = np.random.rand(100)
axs[0, 1].scatter(scatter_x, scatter_y, c=colors, cmap='viridis', alpha=0.6)
axs[0, 1].set_title('Plant Distribution (Random Clusters)')

# --- Plot 3: Bar Chart (Bottom Left) ---
axs[1, 0].bar(categories, values, color=['gray', 'orange', 'blue', 'gold'])
axs[1, 0].set_title('Fuel Type Capacity (Current Year)')

# --- Plot 4: Histogram (Bottom Right) ---
axs[1, 1].hist(dist_data, bins=30, color='tab:green', edgecolor='white')
axs[1, 1].set_title('Efficiency Distribution (Normal Dist)')

# 3. Final Adjustments
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust to fit the suptitle
plt.show()
```
