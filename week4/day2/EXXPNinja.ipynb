```python
import numpy as np

# 1. Create a 4x4 matrix (using random integers 1-10)
m1 = np.random.randint(1, 11, size=(4, 4))

# 2. Extract the diagonal elements
# This returns a 1D array: [a_11, a_22, a_33, a_44]
diag_elements = np.diag(m1)

# 3. Create a diagonal matrix from these elements
# This returns a 4x4 matrix where all non-diagonal elements are 0
diag_matrix = np.diag(diag_elements)

print("Original 4x4 Matrix:\n", m1)
print("\nExtracted Diagonal Elements:", diag_elements)
print("\nNew Diagonal Matrix:\n", diag_matrix)
```

```python

# Exercise 2 : Conditional Array Operations
# Instructions

# Generate a 1D NumPy array with 20 random elements. Replace all elements greater than 0.5 with the value -1.
arr = np.random.random(20)
print(arr)
arr[arr > 0.5] = -1
print(arr)
```

```python
# Exercise 3 : Data Normalization
# Instructions

# Normalize a 1D NumPy array (subtract the mean and divide by the standard deviation).
from sklearn.preprocessing import StandardScaler
data = np.random.randint(1,101,50)
data_reshaped = data.reshape(-1, 1)

# 3. Initialize and fit the scaler
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data_reshaped)

# 4. Flatten back to 1D (optional, for easier viewing)
final_result = standardized_data.flatten()

print(f"Original Data: {data}")
print(f"Standardized Data: {final_result}")
```

```python
# Exercise 4 : Correlation Coefficients
# Instructions

# Given two arrays of data, compute the Pearson correlation coefficient between them using NumPy.
x = np.random.randint(1,101,50)
y = np.random.randint(1,101,50)

matrix = np.corrcoef(x, y)

#Extract the Pearson coefficient from the matrix
pearson_r = matrix[0, 1]

print(f"Pearson Correlation Coefficient: {pearson_r:.2f}")
```
