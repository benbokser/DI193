```python
# Exercise 1 : Geometry
# Instructions

#     Write a class called Circle that receives a radius as an argument (default is 1.0).
#     Write two instance methods to compute perimeter and area.
#     Write a method that prints the geometrical definition of a circle.
pi = 3.14159
class Circle:
    def __init__(self, radius=1.0):
        self.radius = radius
    def perimeter(self):
        return 2*pi*self.radius
    def area(self):
        return pi*(self.radius**2)
    def definition(self):
        print('A circle is an infinite set of points of equal distance from a central point.')

circle1 = Circle(3)
print(circle1.area())
```
